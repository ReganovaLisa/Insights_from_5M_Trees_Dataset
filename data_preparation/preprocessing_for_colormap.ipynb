```python
import pandas as pd
import json
from pandas import json_normalize
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
f = open("../map/data/counties-albers-10m.json")
data = json.load(f)
f.close()
df = json_normalize(data['objects']['states']['geometries'])
```

```python
state_id = df[df.columns[-2:]]
state_id.columns = ['id', 'state']
```

```python

file_path = '../data_raw/'
file_list = os.listdir(file_path)
output_path = '../comparing_categories/data/california_top_five_trees.csv'
df_append = pd.DataFrame()
#append all files together
for file in file_list:
            df_temp = pd.read_csv(file_path  + file, on_bad_lines='skip')
            df_temp['file_name'] = file.split('_')[0]
            df_append = df_append.append(df_temp, ignore_index=True)
```

```python
groups_by_state = df_append.groupby(['state']).count()
result_selected = groups_by_state['city']
result_selected['state'] = result_selected.index
result_selected = result_selected.reset_index()
result_selected = result_selected.iloc[:-1,:]
result_selected.columns = ['state', 'rate']
```

```python
result = result_selected.merge(state_id, how='left', on='state')
```

```python
result.to_csv('../map/data/result_states.csv')
```
